```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Pyplot API (Quick & Easy Way)

```python
plt.figure(figsize=(6, 4))
plt.title("My First Empty Figure")
plt.show()
```

2. Object-Oriented API (Preferred for Complex Plots)

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.set_title("My first Empty Figure")
plt.show()
```

| Aspect        | Pyplot API (`plt.*`)           | OO API (`fig, ax`)            |
|--------------|--------------------------------|------------------------------|
| **Control**   | Implicit (global "current" Axes) | Explicit (you specify Axes)  |
| **Flexibility**| Simple, quick plots           | Complex, multi-plot layouts   |
| **Recommended**| Small scripts                | Larger projects, reusable code|


```python
x = [1, 2, 3, 4, 5, 6]
y = [10, 8, 1, 6, 3, 5]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```
