### Line plots

To show trends over time or an ordered x-axis (e.g., sales over months, temperature over days).

1. Basic Example (One Line)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]     # X-axis values (e.g., Days)
y = [2, 4, 6, 8, 10]    # Y-axis values (e.g., Sales)

fig, ax = plt.subplots()
ax.plot(x, y,linestyle='-', marker='o', color='green', label='Sales')
ax.set(title="Sales Over Days", xlabel="Day", ylabel="Sales")

ax.legend() # Add legend (displays the label of the line)
plt.show()
```

3. Two Lines with Legend

```python
# Data
days = [1, 2, 3, 4, 5]
sales = [2, 4, 6, 8, 10]
profit = [1, 3, 4, 6, 9]

fig, ax = plt.subplots()

ax.plot(days, sales, marker='o', label='sales')
ax.plot(days, profit, marker='s',label='Profit')

ax.set(title="Sales & Profit Over Days", xlabel="Day", ylabel="Value")
ax.legend()
plt.show()
```
