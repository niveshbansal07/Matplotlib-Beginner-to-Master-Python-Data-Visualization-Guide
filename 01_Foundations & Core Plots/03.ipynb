### Scatter Plots
[Ultimate Roadmap →](https://niveshbansal07.github.io/Matplotlib-Roadmap-2-Day-Beginner-to-Master-Python-Data-Visualization-Guide/)

To see the relationship between two variables (e.g., Height vs Weight, Study Hours vs Marks).

1. Basic Example (One Scatter Plot)

```python
import matplotlib.pyplot as plt

# Data
hours = [1, 2, 3, 4, 5, 6, 7]
scores = [35, 40, 50, 65, 70, 75, 90]

fig, ax = plt.subplots()

ax.scatter(hours, scores, s=40, alpha=0.7, color='blue')
ax.grid(True, linestyle=':')

ax.set(title="Study Hours vs Exam Score", xlabel="Hours Studied", ylabel="Score")
```

s=40 → Size of points.

alpha=0.7 → Transparency (0 = invisible, 1 = solid).

ax.grid(True) → Adds background grid, linestyle=':' makes it dotted.

2. Two Groups (Compare Data Sets)

```python
groupA_x = [2, 3, 4, 5, 6]
groupA_y = [20, 25, 30, 35, 40]

groupB_x = [2, 3, 4, 5, 6]
groupB_y = [15, 18, 28, 32, 38]

fig, ax = plt.subplots()

ax.scatter(groupA_x, groupA_y, s=60, alpha=0.7, color='green', label='Group A')
ax.scatter(groupB_x, groupB_y, s=60, alpha=0.7, color='blue', label='Group B')

ax.grid(True, linestyle=':')
ax.set(title='Group Comparison', xlabel='X Value', ylabel='Y Value')

ax.legend()

plt.show()
```
