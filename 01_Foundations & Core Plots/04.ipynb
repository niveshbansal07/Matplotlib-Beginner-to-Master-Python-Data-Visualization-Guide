To compares values across categories using rectangular bars.

1. Vertical Bar Chart (ax.bar)

```python
import matplotlib.pyplot as plt

# Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 15, 7, 12]

fig, ax = plt.subplots()

ax.bar(categories, values, color='skyblue')
ax.set(title='Fruit Sales (Vertical)', xlabel='Fruit', ylabel='Quantity')

ax.set_ylim(0, 20) # Set Y-axis range (0 to 20)

plt.show()
```

2. Horizontal Bar Chart (ax.barh)

```python
# Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 15, 7, 12]

fig, ax = plt.subplots()

ax.barh(categories, values, color='lightgreen')
ax.set(title='Fruit Sales (Horizontal)', xlabel='Quantity', ylabel='Fruit')

ax.set_xlim(0, 20) # Set X-axis range (0 to 20)

plt.show()
```
