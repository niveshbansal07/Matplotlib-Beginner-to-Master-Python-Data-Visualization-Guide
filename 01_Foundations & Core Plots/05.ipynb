### Histograms & Density

To see how data is spread (normal, skewed, outliers).

1. Basic Histogram Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(loc=50, scale=15, size=1000)

fig, ax = plt.subplots()

ax.hist(x, bins=10, edgecolor='black', color='skyblue')
ax.set(title="Hisogram (10 bins)", xlabel="Score", ylabel="Frequency")
plt.show()
```

2. Compare 10 bins vs 30 bins

```python
x = np.random.normal(loc=50, scale=15, size=1000)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.hist(x, bins=10, color='skyblue', edgecolor='black')
ax1.set(title='10 Bins', xlabel='Score', ylabel='Frequency')

ax2.hist(x, bins=30, color='lightgreen', edgecolor='black')
ax2.set(title='30 Bins', xlabel='Score', ylabel='Frequency')

plt.tight_layout()
plt.show()
```

3. Density (Probability Instead of Count)

```python
x = np.random.normal(0, 1, 1000)

fig, ax = plt.subplots()
ax.hist(x, bins=20, density=True, color='orange', alpha=0.7, edgecolor='black')
ax.set(title='Density Histogram', xlabel='Value', ylabel='Probability')
plt.show()
```

density=True â†’ Area under bars = 1 (probability distribution).
