### Box & Violin Plots

**Box Plot**

- Shows spread of data (min, Q1, median, Q3, max).

- Good for comparing distributions across groups.

**Violin Plot**

- Similar to boxplot but also shows data density (like a rotated KDE inside).

1. Boxplot of 3 Groups (Side-by-Side)

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 12, 100)
group3 = np.random.normal(55, 8, 100)

data = [group1, group2, group3]

fig, ax = plt.subplots()
ax.boxplot(data, labels=['group1', 'group2', 'group3'])
ax.set(title='Boxplots of 3 groups', ylabel='Value')

plt.show()
```

2. Violin Plot of Same 3 Groups

```python
fig, ax = plt.subplots()

ax.violinplot(data, showmedians=True) 
ax.set(title='Violin Plot of 3 Groups', xticks=[1,2,3], xticklabels=['Group 1','Group 2','Group 3'], ylabel='Value')

plt.show()

# showmedians=True adds a median line.
```
