**Pie Chart:**

Shows how a whole is divided into parts.

Example:

- If you have 4 fruits:
- Apple 30%, Mango 25%, Orange 25%, Banana 20%.
- A pie chart will show each as a slice of the circle.

**Donut Chart:**

- A Pie Chart with a hole in the middle (looks like a donut).

- Same as a pie, but easier to read and looks cleaner.

- Sometimes you can put a total value or text inside the hole.

1. Pie Chart

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
values = [30, 25, 25, 20]

fig, ax = plt.subplots(figsize=(5, 5))

ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

ax.set_title("Pie Chart Example")

plt.show()
```

2. Donut Chart

```python
labels = ['Apple', 'Mango', 'Orange', 'Banana']
values = [30, 25, 25, 20]

fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

# Add white circle for donut effect
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
ax.add_artist(centre_circle)

ax.set_title("Donut Chart Example")

plt.show()
```

- autopct='%1.1f%%' â†’ Shows percentages on slices.

- startangle=90 â†’ Rotates the start point (optional).

- plt.Circle((0,0), 0.70, fc='white') â†’ Makes a donut by adding a white circle.
