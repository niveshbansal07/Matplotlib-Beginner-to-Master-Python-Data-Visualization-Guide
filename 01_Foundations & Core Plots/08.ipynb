To Points to a specific data point with text and an arrow.

1. Annotate the Maximum Point

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label='sin(x)')

x_max = x[np.argmax(y)]
y_max = np.max(y)

ax.annotate('Max Point',
            xy=(x_max, y_max),             
            xytext=(x_max+1, y_max+.3),
            arrowprops=dict(color='red',arrowstyle='->'))

ax.text(0.5, 0.9, 'Sine Wave', transform=ax.transAxes, fontsize=10, color='blue')

ax.set(title='Annotation Example', xlabel='X-axis', ylabel='Y-axis')
ax.legend()
ax.grid(True, linestyle=':')

plt.show()
```

2. Annotate Both Maximum & Minimum Point

```python
x2 = np.linspace(0, 10, 100)
y2 = np.sin(x2)

fig, ax = plt.subplots()

ax.plot(x,y, label='Sin Wave')

x2_max, y2_max = x2[np.argmax(y2)], np.max(y2)
x2_min, y2_min = x2[np.argmin(y2)], np.min(y2)

ax.annotate('Max Point', xy=[x2_max, y2_max], xytext=(x2_max+1, y2_max+0.2), arrowprops=dict(color='green', arrowstyle='->'))

ax.annotate('Min Point', xy=[x2_min, y2_min], xytext=(x2_min+1, y2_min+0.1), arrowprops=dict(color='red', arrowstyle='->'))

ax.text(0.4, 0.9, 'Sine Wave', transform=ax.transAxes, fontsize=12, color='purple')

ax.grid(True, linestyle=':')
ax.set(title='Annotate Example', xlabel='X-axis', ylabel='Y-axis')
ax.legend()
plt.show()
```
