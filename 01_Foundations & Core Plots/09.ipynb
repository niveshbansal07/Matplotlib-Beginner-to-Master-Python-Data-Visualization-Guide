### Legends, Ticks, Spines
[Ultimate Roadmap â†’](https://niveshbansal07.github.io/Matplotlib-Roadmap-2-Day-Beginner-to-Master-Python-Data-Visualization-Guide/)

1. Legend
- Shows which line/bar/plot corresponds to which data.

2. Ticks
- These are the marks and labels on the X/Y axes.

3. Spines
- The borders around your plot.

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales_2024 = [100, 120, 90, 150, 200]
sales_2025 = [110, 130, 100, 160, 220]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(months, sales_2024, label='Sales 2024', marker='o')
ax.plot(months, sales_2025, label='Sales 2025', marker='s')

# Move legend outside the plot
ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1))

# Rotate X-axis tick labels
ax.set_xticks(range(len(months)))
# ax.set_xticks([0, 1 ,2 , 3, 4])
ax.set_xticklabels(months, rotation=45)

# Keep spines visible (you can also hide them)
for s in ax.spines.values():
    s.set_visible(True)

ax.set(title='Monthly Sales Comparison', xlabel='Month', ylabel='Sales')
ax.grid(True, linestyle=':')

plt.tight_layout()
plt.show()
```
