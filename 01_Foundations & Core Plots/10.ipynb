### Saving Figures
[Ultimate Roadmap â†’](https://niveshbansal07.github.io/Matplotlib-Roadmap-2-Day-Beginner-to-Master-Python-Data-Visualization-Guide/)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label='Sine Wave')
ax.set_title("Saving Figures Example")
ax.legend()

# Adjust layout to avoid cut-off labels
fig.tight_layout()

# Save as PNG (300 DPI = High Quality)
fig.savefig('plot.png', dpi=300, bbox_inches='tight')

# Save as PDF (Vector Format, scalable without losing quality)
fig.savefig('plot.pdf', dpi=300, bbox_inches='tight')

plt.show()
```

- fig.tight_layout() â†’ Adjusts spacing automatically so labels/titles aren't cut off.

- fig.savefig('filename.png', dpi=300) â†’ Saves high-quality image.

- bbox_inches='tight' â†’ Trims extra white space around the figure.

- dpi=300 â†’ Professional print quality.
