- When you want to show multiple plots together.

1. Subplots - Create Multiple Plots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
data = np.random.randn(100)

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Line")
axs[0, 1].scatter(x, np.cos(x))
axs[0, 1].set_title("Scatter")
axs[1, 0].hist(data, bins=10)
axs[1, 0].set_title("Histogram")
axs[1, 1].bar(['A', 'B', 'C'], [5, 3, 7])
axs[1, 1].set_title("Bar")

fig.tight_layout() # To avoid overlap
fig.suptitle("Subplots") #Main Title

# fig.subplots_adjust(wspace=0.5, hspace=0.5) # fig.subplots_adjust(wspace=0.3)

plt.show()
```

2. Layouts - Custom Subplot Arrangements

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])

ax1.plot(x, np.sin(x))
ax1.set_title("Sine")
ax2.plot(x, np.cos(x))
ax2.set_title("Cosine")
ax3.hist(np.random.randn(100), bins=15)
ax3.set_title("Histogram")

fig.suptitle("Custom Layouts")
fig.tight_layout()
plt.show()
```
