### Dual/Secondary Axes

- Dual axes let you plot two different scales (e.g., temperature & rainfall) on one graph for easy comparison.

1. Secondary Y-Axis (twinx)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 13)
temp = np.random.uniform(10, 35, 12)
rain = np.random.uniform(20, 200, 12)

fig, ax1 = plt.subplots(figsize=(8,6))

ax1.plot(x, temp, color='red', marker='o')
ax1.set_xlabel("Month")
ax1.set_ylabel("Temp (Â°C)", color='red')

ax2 = ax1.twinx()  # Secondary Y-axis
ax2.bar(x, rain, alpha=0.5, color='blue')
ax2.set_ylabel("Rain (mm)", color='blue')

plt.title("Temperature vs Rainfall")
plt.show()
```

2. Secondary X-Axis (secondary_xaxis)

```python
import numpy as np
import matplotlib.pyplot as plt

def km_to_miles(x):
    return x * 0.621
def miles_to_km(x):
    return x / 0.621

x = np.linspace(0, 100, 100)
y = np.sqrt(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("Distance (km)")

ax_top = ax.secondary_xaxis('top', functions=(km_to_miles, miles_to_km))
ax_top.set_xlabel("Distance (miles)")

plt.show()
```

- ax.twinx() â†’ Second Y-axis (right side).
- ax.twiny() â†’ Second X-axis (top).
- ax.secondary_xaxis() / ax.secondary_yaxis() â†’ Custom scale conversions.
