- Scales adjust how data is displayed (linear, log, etc.), and limits control which part of data is shown.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 10, 400)
y = np.exp(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

ax1.plot(x, y)
ax1.set_title("Linear Scale")
ax1.set_xlim(0, 12) 

ax2.plot(x, y)
ax2.set_title("Log Y-Scale")
ax2.set_yscale('log')

plt.tight_layout()
plt.show()
```

1. Set Scale (Linear, Log, Symlog)

- ax.set_xscale('log')       # Logarithmic X-axis
- ax.set_yscale('symlog')    # Symmetric Log Y-axis (handles +ve/-ve)

2. Set Limits

- ax.set_xlim(a, b)          # X-axis range
- ax.set_ylim(c, d)          # Y-axis range
