- Colormaps use color to represent data values & colorbars explain the mapping.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(6, 6)

fig, ax = plt.subplots()
# im = ax.imshow(data)
im = ax.imshow(data, cmap='viridis') # default
fig.colorbar(im, ax=ax)

ax.set_title("Heatmap Example")
plt.show()
```

```python
# List all colormaps
plt.colormaps()  
```
