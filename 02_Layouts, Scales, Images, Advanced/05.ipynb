- Used to visualize 2D data (grids, images, intensity maps).

```python
import numpy as np
import matplotlib.pyplot as plt

x = y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2  + Y**2))

fig, ax = plt.subplots()
im = ax.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='magma')

ax.contour(X, Y, Z, colors='white')

fig.colorbar(im, ax=ax)
ax.set_title("Heatmap with Contours")
plt.show()
```
