- Show uncertainty or variability in your data using error bars or confidence bands.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)
err = 0.2 + 0.2*np.sqrt(x)   # variable error

fig, ax = plt.subplots()
ax.plot(x, y, label="Mean")
ax.fill_between(x, y-err, y+err, alpha=0.3, label="95% CI")
ax.legend()
ax.set_title("Line with Confidence Band")
plt.show()
```
