- Format date/time on X-axis for clean, readable charts.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
import numpy as np

dates = [dt.date(2025, m, 1) for m in range(1, 13)]
values = np.random.randint(10, 100, size=12)

fig, ax = plt.subplots()
ax.plot(dates, values, marker='o', label='Sales')

# Format X-axis as 'Jan 2023'
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotate labels automatically
fig.autofmt_xdate()

ax.set(title='Monthly Sales(2025)', ylabel='Sales')
ax.legend()
plt.show()
```
