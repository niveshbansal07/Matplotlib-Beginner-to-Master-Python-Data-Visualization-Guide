- Styling in Matplotlib controls the look and feel of plots (colors, fonts, backgrounds, themes) to make them clear and professional.

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')        # Apply theme
plt.rcParams['font.size'] = 11  # Set font size

fig, ax = plt.subplots()
fig.set_facecolor('#fff')      # Figure background
ax.set_facecolor("#938a8a")    # Plot area background

ax.plot([1, 2, 3], [3, 2, 5], label='Data')
ax.legend()
plt.show()
```
