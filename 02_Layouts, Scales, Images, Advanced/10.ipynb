### Export Like a Pro
[Ultimate Roadmap →](https://niveshbansal07.github.io/Matplotlib-Roadmap-2-Day-Beginner-to-Master-Python-Data-Visualization-Guide/)

- Exporting makes your plots shareable in high quality for reports, web, or print.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 4, 3])

fig.set_dpi(150)                        
fig.savefig('figure.svg')               
fig.savefig('figure.png', dpi=300, 
            transparent=True)          
plt.show()
```
